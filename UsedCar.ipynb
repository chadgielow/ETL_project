```python
#Dependancies
import pandas as pd
import numpy as np
```

```python
#Csv to DataFrame
read_csv = "craigslistVehicles.csv"
cars = pd.read_csv(read_csv)
cars.head()
```

```python
cleaned_data = cars[["manufacturer","make","year", "price"]]
cleaned_data.head()
```

```python
cleaned_data.astype({'year':'str'}).dtypes
```

```python
cleaned_data.year = cleaned_data['year'].astype(str).str[0:4]

cleaned_data.head()
```

```python
cleaner = cleaned_data.dropna()
cleanest = cleaner.set_index('manufacturer')
cleanest
```

```python
# Export to csv
cleanest.to_csv("Data/cleaned_used_car.csv")
```
